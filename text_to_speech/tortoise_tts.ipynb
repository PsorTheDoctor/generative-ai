```python
import torch
import torchaudio
import torch.nn as nn
import torch.nn.functional as F
from IPython.display import Audio
from tortoise.api import TextToSpeech
from tortoise.utils.audio import load_audio, load_voice, load_voices

tts = TextToSpeech()  # Downloads all the model used by Tortoise from HF hub.
```

```python
text = 'Hello world!'
preset = 'fast'  # Options: "ultra_fast", "fast", "standard" (dafault), "high_quality"
```

```python
%ls tortoise/voices
```

```python
Audio('tortoise/voices/tom/1.wav')
```

```python
voice = 'tom'
voice_samples, conditioning_latents = load_voice(voice)
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=conditioning_latents,
                          preset=preset)
torchaudio.save('generated.wav', gen.squeeze(0).cpu(), 24000)
Audio('generated.wav')
```

```python
# Random voice
gen = tts.tts_with_preset(text, voice_samples=None, conditioning_latents=None, preset=preset)
torchaudio.save('generated2.wav', gen.squeeze(0).cpu(), 24000)
Audio('generated2.wav')
```

```python
# Combining voices
voice_samples, conditioning_latents = load_voices(['pat', 'william'])
gen = tts.tts_with_preset(text, voice_samples=voice_samples, conditioning_latents=None, 
                          preset=preset)
torchaudio.save('generated3.wav', gen.squeeze(0).cpu(), 24000)
Audio('generated3.wav')
```
