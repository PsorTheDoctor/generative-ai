```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio

preload_models()
```

```python
prompt = '♪♪ Parostatkiem w piękny rejs, ♪♪ statkiem na parę w piękny rejs ♪♪'
```

```python
audio_array = generate_audio(
    prompt, history_prompt=f'v2/pl_speaker_4'
)
write_wav(f'result.wav', SAMPLE_RATE, audio_array)

Audio(audio_array, rate=SAMPLE_RATE)
```

```python
# Test all the Polish speakers
for i in range(10):
  audio_array = generate_audio(
      prompt, history_prompt=f'v2/pl_speaker_{i}'
  )
  write_wav(f'result{i}.wav', SAMPLE_RATE, audio_array)
```
