```python
from PIL import Image
from diffusers import StableDiffusionUpscalePipeline
import torch

modelId = "stabilityai/stable-diffusion-x4-upscaler"
pipeline = StableDiffusionUpscalePipeline.from_pretrained(modelId, revision="fp16", torch_dtype=torch.float16)
pipeline = pipeline.to("cuda")
```

```python
# You need to upload the file, eg. using left sidebar in Google Colab
lowResImg = Image.open('adam.jpg').convert('RGB')
lowResImg = lowResImg.resize((128, 128))

prompt = 'a photo of a young man'

upscaledImg = pipeline(prompt=prompt, image=lowResImg).images[0]
upscaledImg.save('result.png')
upscaledImg
```
